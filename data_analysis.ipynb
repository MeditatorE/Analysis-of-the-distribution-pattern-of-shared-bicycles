```python
import matplotlib.pyplot as plt
```

```python
x_Partially_cloudy=df_date_Partially_cloudy['datetime'].tolist()
y_Partially_cloudy=df_date_Partially_cloudy['rental quantity'].tolist()
x_Rain=df_date_Rain['datetime'].tolist()
y_Rain=df_date_Rain['rental quantity'].tolist()
x_Overcast=df_date_Overcast['datetime'].tolist()
y_Overcast=df_date_Overcast['rental quantity'].tolist()
x_Clear=df_date_Clear['datetime'].tolist()
y_Clear=df_date_Clear['rental quantity'].tolist()
x_Snow=df_date_Snow['datetime'].tolist()
y_Snow=df_date_Snow['rental quantity'].tolist()
x_Freezing_Drizzle_Freezing_Rain=df_date_Freezing_Drizzle_Freezing_Rain['datetime'].tolist()
y_Freezing_Drizzle_Freezing_Rain=df_date_Freezing_Drizzle_Freezing_Rain['rental quantity'].tolist()
x_Ice=df_date_Ice['datetime'].tolist()
y_Ice=df_date_Ice['rental quantity'].tolist()
```

```python
plt.figure(figsize=(50, 30))
plt.plot(x_Partially_cloudy, y_Partially_cloudy, label='Cloudy',linewidth=5)
plt.plot(x_Rain, y_Rain, label='Rain',linewidth=5)
plt.plot(x_Overcast, y_Overcast, label='Overcast',linewidth=5)
plt.plot(x_Clear, y_Clear, label='Clear',linewidth=5)
plt.plot(x_Snow, y_Snow, label='Snow',linewidth=5)
plt.plot(x_Freezing_Drizzle_Freezing_Rain, y_Freezing_Drizzle_Freezing_Rain, label='Freezing Drizzle/Freezing Rain',linewidth=5)
plt.plot(x_Ice, y_Ice, label='Ice',linewidth=5)
plt.xlabel('Date',fontsize=50)
plt.ylabel('Rental quantity',fontsize=50)
plt.legend(fontsize=45)
plt.xticks(rotation=45,fontsize=50)
plt.yticks(fontsize=50)
plt.tight_layout()

# 显示图形
plt.show()
```

```python
x_total=df_date_rent_num['datetime'].tolist()
y_total=df_date_rent_num['rental quantity'].tolist()
```

```python
plt.figure(figsize=(50, 30))
plt.plot(x_total, y_total, label='total',linewidth=5)
plt.xlabel('Date',fontsize=50)
plt.ylabel('Rental quantity',fontsize=50)
plt.legend(fontsize=45)
plt.xticks(rotation=45,fontsize=50)
plt.yticks(fontsize=50)
plt.tight_layout()

# 显示图形
plt.show()
```

**Note:** 

The above figure respectively counts the changes in the number of leases in different weather conditions and the changes in the total number of leases in the past four years. Two conclusions can be drawn.

1. The weather has an impact on the number of leases, but the impact is limited. Snow has the greatest impact on the number of leases in different weather conditions, while cloudy, overcast and even rainy days have little impact on the number of leases. The greater impact is that during holidays, we can see that the data has obvious cyclical fluctuations. The trough is in January every year, which may be due to the New Year's Day holiday and people don't need to go to work. The peak is in July-September every year, which may be due to the school holidays and students going out to play.

2. The number of leases has been increasing in the past four years. Next, we will use Gaussian smoothing to try to smooth the curve on the data, so that it is easier to calculate the annual growth rate.

```python
from scipy.ndimage import gaussian_filter
```

```python
y_total_smoothed = gaussian_filter(y_total, sigma=1000)
plt.figure(figsize=(50, 30))
plt.plot(x_total, y_total_smoothed, label='total',linewidth=5)
plt.xlabel('Date',fontsize=50)
plt.ylabel('Rental quantity',fontsize=50)
plt.legend(fontsize=45)
plt.xticks(rotation=45,fontsize=50)
plt.yticks(fontsize=50)
plt.tight_layout()
```

After setting sigma to 1000, the curve is almost smoothed into a straight line. In this state, noise and fluctuations are completely removed, and the growth rate can be obtained by calculating the slope.
